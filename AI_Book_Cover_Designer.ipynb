```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
from IPython.display import display

model_id = "stabilityai/stable-diffusion-2-1"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

def generate_book_cover(title, summary):
    prompt = f"Book cover design for a story titled '{title}'. {summary}. Artistic, detailed, professional book cover illustration."
    print("ðŸŽ¨ Generating cover for:", title)
    image = pipe(prompt).images[0]
    return image

book_title = input("Enter the book title: ")
book_summary = input("Enter a short summary: ")

cover_image = generate_book_cover(book_title, book_summary)
display(cover_image)

```
